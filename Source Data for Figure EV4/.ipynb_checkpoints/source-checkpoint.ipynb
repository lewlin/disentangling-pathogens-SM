```python
import os, sys
import numpy as np
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
import matplotlib as mpl
import statsmodels.api as sm
from statsmodels.distributions import empirical_distribution
from statsmodels.stats.outliers_influence import summary_table
import warnings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
# import lib
```

****
## Analysis of data from Levine's lab

Load dataset

```python
data_path = os.path.join('./../levine/', 'Fig 2C. N2.mat')
data = scipy.io.loadmat(data_path)
ts = data['a'][:, 0]
ys = data['a'][:, 1] / 100
es = np.sqrt(data['a'][:, 1]) / 100
ys_log = np.log10(ys)
mask =  ts < 55
ts2 = ts[mask]
ys2 = ys[mask]
es2 = es[mask]
ys2_log = ys_log[mask]
```

Fit data

```python
fit_low = 38
fit_high = 55
mask = [fit_low < t < fit_high for t in ts2]
ts3 = ts2[mask]
ys3 = ys2[mask]
es3 = es2[mask]
ys3_log = ys2_log[mask]
X = sm.add_constant(ts3)
Y = ys3_log
ols = sm.OLS(Y, X).fit()
ols.summary()
# ts2_er = ts2
# ts3_er = ts3
# ys2_er = ys2
# ys3_er = ys2
```

Display data / fit and residual plot.

```python
fig1, axs = plt.subplots(ncols=2, figsize=(15,4))
axs[0].errorbar(ts2, ys2, es2, alpha=.5);
axs[0].set_xlabel('Time (hr)')
axs[0].set_ylabel('Fraction surviving')
axs[0].set_yscale('log')
axs[0].set_title('C. elegans/PA14 system (from Levine 2016)');
mpl.table.table(
            #colLabels=type(self)._col_labels,
            #colColours=np.asarray([type(self)._color, type(self)._color]),
            cellText=[['delta:', '0.06 1/h'], ['# obs. in fit', '16'],  ['# worms in exp', '100'],  ['Model', 'OLS']],
#             cellText=[['delta:', '0.05 1/h'], ['R^2', '97%'], ['F (p-value)', '426 (1e-9)']],
            loc='lower left',
            colWidths=[0.2,0.2],
            # colWidths=[type(self)._col_width] * 2,
            ax=axs[0],
            zorder=3,
            colLoc='right',
            # colLoc=type(self)._col_loc,
            rowLoc='right',
        )
axs[0].plot(ts3, np.power(10, ols.predict()), 'k-', lw=.5);
axs[1].plot(ts3, ols.resid, 'o')
axs[1].set_xlabel('Time (h)')
axs[1].set_title('Residual plot')
axs[1].axhline(0, color='k', lw=.5)
axs[1].set_ylabel('Residual');
ax_levine = axs[0]
```

## Fontana data

Load data

```python
fontana_path = os.path.join('./../fontana/', '2010_11_23_ten_n2=survival.csv')
df = pd.read_csv(fontana_path)
df = df[df.Technique == 'Lifespan Machine']
data_fn = df['Age at Death (d) Raw']
ecdf = empirical_distribution.ECDF(data_fn)
ts_fn = ecdf.x * 24
ys_fn = 1-ecdf.y
es_fn = np.sqrt(ys_fn * len(ecdf.y)) / len(ecdf.y)
mask = [t < 18 * 24 for t in ts_fn]
ts2_fn = ts_fn[mask]
ys2_fn = ys_fn[mask]
es2_fn = es_fn[mask]
```

First fit

```python
mask = [(300 < t) and (t < 350) for t in ts2_fn]
ts3_fn = ts2_fn[mask]
ys3_fn = ys2_fn[mask]
es3_fn = es2_fn[mask]
X_1 = sm.add_constant(ts3_fn)
Y_1 = np.log10(ys3_fn)
ols_fn_1 = sm.OLS(Y_1, X_1).fit()
ols_fn_1.summary()
```

```python
mask = [(370 < t) and (t < 400) for t in ts2_fn]
ts4_fn = ts2_fn[mask]
ys4_fn = ys2_fn[mask]
es4_fn = es2_fn[mask]
X_2 = sm.add_constant(ts4_fn)
Y_2 = np.log10(ys4_fn)
ols_fn_2 = sm.OLS(Y_2, X_2).fit()
ols_fn_2.summary()
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(15, 5), sharey=True)

## Levine data
axs[0].errorbar(ts2, ys2, es2, alpha=.5);
axs[0].set_xlabel('Time (hr)')
axs[0].set_ylabel('Fraction surviving')
axs[0].set_yscale('log')
axs[0].set_title('C. elegans/PA14 system (from Levine 2016)');
mpl.table.table(
            cellText=[['delta:', '0.06 1/h'], ['# worms in exp', '100'],  ['Model', 'OLS']],
            loc='lower left',
            colWidths=[0.2,0.2],
            ax=axs[0],
            zorder=3,
            colLoc='right',
            rowLoc='right',
        )
axs[0].plot(ts3, np.power(10, ols.predict()), 'k-', lw=.5);

## Fontana data
axs[1].errorbar(ts2_fn, ys2_fn, es2_fn, alpha=.5);
ts_fit_1 = [250, 400] 
ys_fit_1 = np.power(10, ols_fn_1.predict(exog=[[1, 250],[1, 400]]))
axs[1].plot(ts_fit_1, ys_fit_1, 'k', lw=.5)
ts_fit_2 = [320, 450] 
ys_fit_2 = np.power(10, ols_fn_2.predict(exog=[[1, 320],[1, 450]]))
axs[1].plot(ts_fit_2, ys_fit_2, 'k', lw=.5)
axs[1].set_xlabel('Time (hr)')
axs[1].set_title('C. elegans aging survival (from [Fontana 2013])');
mpl.table.table(
            cellText=[['delta_1:', '0.010 1/hr'], ['delta_2:', '0.018 1/hr'], ['# worms', '3390']],
            loc='lower left',
            colWidths=[0.2,0.2],
            ax=axs[1],
            zorder=3,
            colLoc='right',
            rowLoc='right',
        );
axs[1].set_yscale('log')

fig.tight_layout()
```
